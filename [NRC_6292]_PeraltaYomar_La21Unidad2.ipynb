**ERROR 1 ! - Indentacion incorrecta del dataset**

```python
'''Correccion del error numero 1 tratandose de la separacion de los diferentes registros de los campos de todo el dataset'''
df= pd.read_csv('data_act_01.csv', ';')
df
```

**ERROR** **2** ! - **Datos** **extensos**

```python
'''Analisis de los datos del campo AddressType incluyendo registros nulos si los tuviera'''
print(df['AddressType'].value_counts(dropna=False))
```

```python
'''Cambio de datos por string por datos simples del campo AddressType para ahorro de memoria '''
d1 = {'Premise Address': 0, 'Intersection': 1, 'Common Location': 2, 'Geo-Override': 3, '1': 4, 'Intersectioon': 5}
df['AddressType'] = df['AddressType'].map(d1)
```

```python
print(df['AddressType'].value_counts(dropna=False))
```

```python
df
```

```python
'''Visualizacion general del dataset con los cambios ya realizados'''
df.head()
```

```python
'''Detalle de los registros nulos que exitan en cada uno de los campos 
despues de los cambios que se realicen, que en este caso son City, State y Range'''
df.isnull().sum()
```

**ERROR 3 ! - Datos faltantes en el campo City**

```python
import numpy as np
```

```python
''' Como se puede visualizar en el campo de City hacen falta registros, 
entonces para dicho error lo corregimos con su eliminacion''' 
df.isnull().sum()
```

```python
''' Analisis del campo en el cual existen 321 registros faltantes (Nan)'''
print(df['City'].value_counts(dropna=False))
```

```python
''' Impresion de los registros de la columna City y Address 
para la verificacion de los datos nulos que ubicacion tienen '''
df.loc[:, ['City', 'Address']]
```

```python
'''Cambio del registro tipo NaN en el campo City dandole valor de 0'''
df[['City']] = df[['City']].fillna(value=0)
```

```python
print(df['City'].value_counts())
```

```python
df.groupby('City').count()
```

```python
'''Verificacion de la limpieza'''
df.isnull().sum()
```



**ERROR 4 ! - Columna "Range" sin valores ni registros validos** 

```python
'''Eliminacion de la columna "Range" ya que en la misma no existen datos validos'''
df.drop(['Range'], axis='columns', inplace=True)
```

```python
''' Impresion de los datos ya sin la columna Range'''
df
```

```python
'''Verificacion de la limpieza'''
df.isnull().sum()
```

```python
df.groupby('City').count()
```

**ERROR 5 ! Redundancia de mayusculas y minusculas**

```python
'''En el campo de Ciudad  existen registros tipo SAN FRANCISCO, S y San Francisco'''
df['City']= df['City'].replace('San Francisco','SAN FRANCISCO')
```

```python
'''Cambio del registro tipo "S" a "SAN FRANCISCO"'''
df['City']= df['City'].replace(' S', 'SAN FRANCISCO')
```

```python
df['City']= df['City'].replace(0, 'SAN FRANCISCO')
```

```python
df.groupby('City').count()
```

```python
'''Sumatoria de los registros faltantes en el dataset'''
df.isnull().sum()
```

**ERROR 6 ! Valores nulos en el campo State**

```python
''' El campo "AgencyId" posee 3 registros invalidos '''
df[['State']] = df[['State']].fillna(value=0)
```

```python
df
```

```python
print(df['State'].value_counts(dropna=False))
```

```python
df.groupby('State').count()
```

```python
df['State']= df['State'].replace('CA', '1')
```

```python
df.groupby('State').count()
```

```python

'''Verficiacion de la limpieza de los registros nulos en el dataset'''
df.isnull().sum()
```

**ERROR ! - Tipo de registro de una columna ya limpada**

```python
'''Cambio del campo State de tipo object a tipo int'''
df['State']=pd.to_numeric(df['State'], errors='coerce')
```

```python
df.info()
```

**ERROR  ! - Dato numerico incorrecto en campo  Disposition**

```python
'''Visualizacion de los tipos de registros que existen en el campo Disposition'''
print(df['Disposition'].value_counts(dropna=False)) 
```

**ERROR ! - Registro numerico en campo no correspondiente**

```python
'''Cambio del registro "Not Recorded" a NR'''
df['Disposition']= df['Disposition'].replace('Not recorded', 'NR')
```

```python
'''Cambio del registro "22" a ABC'''
df['Disposition']= df['Disposition'].replace('22', 'ABC')
```

```python
print(df['Disposition'].value_counts(dropna=False)) 
```

**ERROR ! - Formato de las fechas y horas del CallDateTime**

```python
'''Con la ayuda del comando datetime se pudo cambiar el formato de la horay fecha del registro'''
df['CallDateTime'] = pd.to_datetime(df['CallDateTime'])
df
```

**ERROR ! - Formato de la fecha inconsistente**

```python
'''Cambio del formato de la hora en el campo OffenseDate'''
df['OffenseDate'] = pd.to_datetime(df['OffenseDate'])
df
```
